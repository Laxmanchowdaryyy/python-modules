```python
import matplotlib.pyplot as plt

import numpy as np

y=np.array([120,80,100,60])

x=("Python","chemistry","English","Maths")

mycolors=('b','orange','green','red')

myexplode=(0.2,0,0,0)

plt.pie(y,labels=x, colors=mycolors, explode=myexplode)

plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np

y=np.array([120,80,100,60])

x=(["cse","mech","civil","ece"])

mycolors=('blue','orange','green','red')

plt.pie(y ,labels=x,colors=mycolors)

plt.legend( title="for subjects")

plt.show()
```
